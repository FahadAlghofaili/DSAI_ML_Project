#Saudi Supply Chain Dataset – Description
This dataset represents a simulated supply chain system in Saudi Arabia. It includes information about product demand, inventory levels, shipments, logistics operations, and various risk factors. The dataset is designed to support multiple machine learning tasks such as classification, regression, time series forecasting, and clustering.

Each row corresponds to a supply chain transaction involving a specific product and shipment.

Link: https://www.kaggle.com/datasets/programmer3/saudi-supply-chain-dataset/data

-----

##1. Time and Product Information
*   date: The date of the recorded transaction. It is useful for time series analysis and tracking trends over time.

*   product_id: A unique identifier assigned to each product.


---


##2. Demand and Inventory

*   historical_demand: The actual observed demand for the product.
*   forecasted_demand_prev: The previously predicted demand value.
*   inventory_level: The current quantity of stock available.

*   order_quantity: The quantity ordered to replenish inventory.
---

##3. Shipping and Logistics


*   shipment_id: A unique identifier for each shipment.
*   origin: The starting location of the shipment.
*   destination: The delivery location.
*   distance_km: The distance between origin and destination in kilometers.
*   transport_mode: The mode of transportation (e.g., air, sea, road).
*   transit_time: The time taken for the shipment to travel from origin to destination.


*   delay_hours: The delay duration in hours, if any.
---

##4. Operational Performance


*   lead_time_days: The number of days required to fulfill an order from supplier to delivery.

*   shipment_status: The current status of the shipment (e.g., delivered, delayed).
*   fuel_cost: The cost associated with fuel used during transportation.
*   stockout_flag: Indicates whether a stockout occurred (0 = No, 1 = Yes).


---

##5. Risk Factors


*   supplier_risk_score: A score indicating the level of risk associated with the supplier.

*   weather_disruption_flag: Indicates whether weather conditions caused a disruption (0 = No, 1 = Yes).
*   political_risk_index: A measure of political instability that may impact the supply chain.
*   labor_strike_flag: Indicates whether labor strikes occurred (0 = No, 1 = Yes).
*   demand_volatility_index: A measure of how much demand fluctuates over time.
*   resilience_score: A score representing how resilient the supply chain is to disruptions.
*   at_risk_shipment: Indicates whether a shipment is considered at risk (0 = No, 1 = Yes).
---
