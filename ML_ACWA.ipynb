# Data Cleaning & Preparation

```python
import pandas as pd
import os

file_path = '/content/drive/MyDrive/Colab Notebooks/ACWA Power 2 - Data Recording Table Template.csv'

if os.path.exists(file_path):
    # Load skipping the very top merged title
    df_raw = pd.read_csv(file_path, skiprows=1)

    # Promote the first row of data to be the Header
    # We use .str.replace and .str.strip correctly here
    new_header = df_raw.iloc[0].astype(str).str.replace('\n', ' ').str.strip()
    df_raw.columns = new_header

    # Drop the metadata rows (Names, Units, Ranges, Relations)
    # Usually the first 4 rows are non-data
    df = df_raw.iloc[4:].copy()

    # 4. Final Cleanup
    df = df.reset_index(drop=True)
    # Remove columns that have no name (Unnamed)
    df = df.loc[:, ~df.columns.str.contains('nan|Unnamed', case=False)]

    # 5. Convert everything to numbers for your calculations
    df = df.apply(pd.to_numeric, errors='coerce')

    print("Success! Data is cleaned, named, and numeric.")
    display(df.head())
else:
    print(f"File not found. Double check the path: {file_path}")
```
